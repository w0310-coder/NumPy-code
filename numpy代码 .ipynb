```python
# 此代码演示了np.where单条件判断，按需替换数组内元素
arr = np.array([1,2,3,4,5])
print(np.where(arr<3, arr, 0))
print(np.where(arr<3,1,0))
```

```python
# 此代码演示了嵌套np.where，实现成绩多档位分段评级
score = np.random.randint(30,100,20)
print(score)
print(np.where(score>=60,'及格','不及格'))
print(np.where(
    score<60,'不及格',np.where(
    score<80,'良好','优秀'
)))
```

```python
# 此代码演示了np.select多条件批量匹配映射结果
score = np.random.randint(30,100,20)
print(np.select([score>80,(score>=60) & (score<=80),score<60],['优秀','良好','不及格']))
```

```python
# 此代码演示了固定随机种子、生成指定范围随机整数数组、并对数组升序排序
np.random.seed(0)
arr = np.random.randint(1,100,20)
print(arr)
print(np.sort(arr))
```
